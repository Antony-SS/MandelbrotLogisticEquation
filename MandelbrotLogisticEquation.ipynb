```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.pyplot import figure
```

```python
# Going to attempt to create the Mandelbrot graph using np

# first we start with the logistic equation

x = np.linspace(0,100, num=100)


# xn+1 = r*xn*(1-xn) where xn is a percentage of the max population 1.  This is a quadratic equation and creates a negative feedback loop.

# This logistic equation should tends towards a constant value regardless of starting population.  Will call this equilibrium population

currentPop = []

currentPop.append(.30) # setting initial condition,

r = 2.5 # growth rate, also chosen arbitrarily



for i in range (1, len(x)):
    toAppend = r*currentPop[i-1]*(1-currentPop[i-1])
    currentPop.append(toAppend)


# Graphing

figure(figsize=(16,8))
plt.plot(x, currentPop)
plt.xlabel("Time Units")
plt.ylabel("Population")


plt.show()

# As you can see the equilbrium population here is around 60% of the max population

# But what happens if we change the growth rate . . . 
```

```python

currentPop = []

currentPop.append(.30) #setting initial condition,

r = .95 # growth rate, also chosen arbitrarily

for i in range (1, len(x)):
    toAppend = r*currentPop[i-1]*(1-currentPop[i-1])
    currentPop.append(toAppend)


# Graphing

figure(figsize=(16,8))

plt.plot(x, currentPop)
plt.xlabel("Time Units")
plt.ylabel("Population")

plt.show()

# When the growth rate is below 1 the population approaches 0, which makes sense (population just shrinking at each interval)
```

```python
# Let us explore what the population will approach with different growth values r

# We're gonna see a pretty cool phenomenom

r = np.linspace(0, 3.0, num=100)

x = np.linspace(0,100, num=1000)

equilPops = []

for i in range(0, len(r)):
    currentPop = []
    currentPop.append(.30) #setting initial condition,
    # growth rate will now change

    for j in range (1, len(x)):
        toAppend = r[i]*currentPop[j-1]*(1-currentPop[j-1])
        currentPop.append(toAppend)
    
    equilPop = currentPop[-1]
    equilPops.append(equilPop)


figure(figsize=(16,8))
plt.scatter(r, equilPops)
plt.xlabel("Growth Rate (r)")
plt.ylabel("Equilibrium Pop")


# Any r value below 1, our population settles on 0.  As we increase r linearly, equilbrium population also increases but in a quadratic fashion
```

```python
# Looking normal . . . 

# Now looks what happens when we increase growth rate past factor of 3

from matplotlib.pyplot import figure

r = np.linspace(0, 3.6, num=200)

x = np.linspace(0,100, num=1000)

equilPops = []

for i in range(0, len(r)):
    currentPop = []
    currentPop.append(.30) #setting initial condition,
    # growth rate will now change

    for j in range (1, len(x)):
        toAppend = r[i]*currentPop[j-1]*(1-currentPop[j-1])
        currentPop.append(toAppend)
    
    equilPop = currentPop[-1]
    equilPops.append(equilPop)


figure(figsize=(16,12))
plt.scatter(r, equilPops)
plt.autoscale()
plt.xlabel("Growth Rate (r)")
plt.ylabel("Equilibrium Pop")
```

```python
# We start alternating between equilbrium population values until we reach value, the graph seems to be splitting

# Lets increase r even more

from matplotlib.pyplot import figure

r = np.linspace(0, 4, num=10000)

x = np.linspace(0,100, num=1000)

equilPops = []

for i in range(0, len(r)):
    currentPop = []
    currentPop.append(.30) #setting initial condition,
    # growth rate will now change

    for j in range (1, len(x)):
        toAppend = r[i]*currentPop[j-1]*(1-currentPop[j-1])
        currentPop.append(toAppend)
    
    equilPop = currentPop[-1]
    equilPops.append(equilPop)


figure(figsize=(16,12))
plt.scatter(r, equilPops)
plt.autoscale()
plt.xlabel("Growth Rate (r)")
plt.ylabel("Equilibrium Pop")
```

```python
# Complete Chaos, but the equlibrium values are essentially constantly alternating between branches.  These branches repeat as the growth rate increases.

# This is actually why the mandelbrot group statbilizes between certain values.  

# Before r = 3, the population will settle on a single value.  

# Between r = 3 and r ~= 3.3 the population will alternate between two branches.

# Beyond r = 3.3 that population will alternate between 4 branches

# Then 8 . . . . 

# You get the idea . . . . 

# Really interesting concept, would love to learn more about this one day, alas don't have time right now.
```
